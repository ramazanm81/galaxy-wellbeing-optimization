```python
# Standardize column names
# - remove leading/trailing spaces
# - convert to lowercase
# - replace spaces with underscores
df.columns = (df.columns.str.lower().str.strip().str.replace(':', '_'))
df.columns
```

```python
perentage_of_loss = 0.91
columns_to_drop = df.columns[df.isna().mean() > perentage_of_loss]

dropped_data = df.drop(columns_to_drop, axis=1)
dropped_data
```

```python
dropped_data.shape
```

```python
#percentage of missing values
dropped_data.isna().mean(axis = 1)
```

```python
#unique values in missing values' percentage
dropped_data.isna().mean(axis = 1).unique()
```

```python
# recieve only rows which percent of nan < 0.6
dropped_data = dropped_data[dropped_data.isna().mean(axis = 1) < 0.6]
dropped_data
```

```python
print("Final shape:", dropped_data.shape)
print("Remaining missing values:", dropped_data.isna().sum().sum())
print("Duplicated rows:", dropped_data.duplicated().sum())
```

```python
#named columns
dropped_data = dropped_data.rename(columns={
    "unnamed_ 4": "monthly_label",
    "unnamed_ 5": "monthly_price",
    "unnamed_ 6": "monthly_unit",
    "unnamed_ 8": "hourly_label",
    "unnamed_ 9": "hourly_price",
    "unnamed_ 10": "hourly_unit",
})
```

```python
dropped_data
```

```python
#i should separate our data on monthly and hourly dataset, because we look differ betweek month and hour
monthly_df = dropped_data[
    dropped_data["monthly_label"].isin([
        "January","February","March","April","May","June",
        "July","August","September","October","November","December"
    ])
][["monthly_label", "monthly_price", "monthly_unit"]]

monthly_df
```

```python
monthly_df.to_excel('monthly_dc.xlsx', index = False)
```

```python
hourly_df = dropped_data[
    dropped_data["hourly_label"].notna()
][["hourly_label", "hourly_price", "hourly_unit"]]

hourly_df
```

```python
hourly_df.to_excel('hourly_dc.xlsx', index = False)
```

```python
monthly_df['monthly_price'] = pd.to_numeric(monthly_df['monthly_price'])
```

```python
hourly_df['hourly_price'] = pd.to_numeric(hourly_df['hourly_price'])
```

```python
monthly_df.isna().sum(), hourly_df.isna().sum()
```

# Conclusion

**The original dataset contained two distinct pricing tables stored in a single Excel sheet.
Due to this structure, a large number of missing values and unnamed columns were present.
The data was cleaned by separating it into logical monthly and hourly pricing tables.
This approach preserved the original information while improving clarity and usability.**
