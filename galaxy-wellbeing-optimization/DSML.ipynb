```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor

X = planets_train.drop(columns=['y'])
y = planets_train['y']


cat_cols = X.select_dtypes(include='object').columns
num_cols = X.select_dtypes(exclude='object').columns

print("Категориальные:", list(cat_cols))
print("Числовые:", list(num_cols))

num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_cols),
        ('cat', cat_transformer, cat_cols)
    ]
)

model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        n_jobs=-1
    ))
])
```

```python
model.fit(X, y)
pred = model.predict(planets_test[X.columns])
```

```python
pred
```

```python
from sklearn.model_selection import cross_val_score
rmse_scores = cross_val_score(
    model,
    X,
    y,
    cv=5,
    scoring='neg_root_mean_squared_error'
)

print("RMSE по фолдам:", -rmse_scores)
print("Средний RMSE:", -rmse_scores.mean())
```

```python
opt_df['existence_expectancy_index'] = opt_df['existence_expectancy_index'].fillna(opt_df['existence_expectancy_index'].median())
```

```python
low_mask = opt_df['existence_expectancy_index'] < 0.7
low_df = opt_df[low_mask]
low_df
```

```python
Total_energy = 50000
min_low_energy = 0.10 * Total_energy
max_per_galaxy = 100

```

```python
opt_df[['pred', 'opt_pred']].head()

```

```python
opt_df.columns
```

```python
opt_df['opt_pred'] = 0.0
```

```python
opt_df[['pred', 'opt_pred']].head()
```

```python
import numpy as np

opt_df['potential'] = -np.log(opt_df['pred'] + 0.01) + 3

opt_df['score'] = opt_df['potential'] ** 2
```

```python
min_low_energy = 0.10 * Total_energy  # 5000
allocated_low = 0
energy_left = Total_energy
```

```python
for idx, row in low_df.iterrows():
    if allocated_low >= min_low_energy:
        break

    give = min(max_per_galaxy, min_low_energy - allocated_low)

    opt_df.at[idx, 'opt_pred'] = give

    allocated_low += give
    energy_left -= give
```

```python
rest_df = opt_df[opt_df['opt_pred'] == 0] \
            .sort_values('score', ascending=False)

for idx, row in rest_df.iterrows():
    if energy_left <= 0:
        break

    give = min(max_per_galaxy, energy_left)

    opt_df.at[idx, 'opt_pred'] = give
    energy_left -= give
```

```python
print("Всего энергии:", opt_df['opt_pred'].sum())
print(
    "Доля <0.7:",
    opt_df.loc[low_mask, 'opt_pred'].sum() / Total_energy
)
print("Максимум:", opt_df['opt_pred'].max())
print("Минимум:", opt_df['opt_pred'].min())
```
